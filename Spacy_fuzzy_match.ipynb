```python
import pandas as pd
import spacy
```

```python
df = pd.read_csv('Resources/news.csv')
```

```python
df
```

```python
nlp = spacy.load('en_core_web_sm')
```

```python
def get_names(text):
    doc = nlp(text)
    
    names = ", ".join([ent.text for ent in doc.ents if ent.label_ == 'ORG'])
    
    return names
    #print(names)
```

```python
df['Orgs'] = df['title'].apply(get_names)
```

```python
company_names = pd.read_csv('clean_company_names.csv')
```

```python
company_name_list = list(company_names['Security Name'])
```

```python
df_2 = df.loc[(df['Orgs'] != '') & (df['date'] != np.nan)]
```
